0/1 Knapsack Problem

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]
  
  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

Task 1: Modify the three techniques to include additional criterion in the knapsack problems

```python
#type your code here
#Recursion
def rec_knapSack(w, v, wt, vol, val, n):
    if n == 0 or w == 0 or v == 0:
        return 0
    if wt[n-1] > w or vol[n-1] > v:
        return rec_knapSack(w, v, wt, vol, val, n-1)
    return max(
        val[n-1] + rec_knapSack(w - wt[n-1], v - vol[n-1], wt, vol, val, n-1),
        rec_knapSack(w, v, wt, vol, val, n-1)
    )

# Test
val = [60, 100, 120]
wt = [10, 20, 30]
vol = [1, 2, 3]
W, V = 50, 5
n = len(val)

print("Recursive:", rec_knapSack(W, V, wt, vol, val, n))



#Dynamic
def DP_knapSack(w, v, wt, vol, val, n):
    table = [[[0]*(v+1) for _ in range(w+1)] for _ in range(n+1)]

    for i in range(1, n+1):
        for j in range(w+1):
            for k in range(v+1):
                if wt[i-1] <= j and vol[i-1] <= k:
                    table[i][j][k] = max(
                        val[i-1] + table[i-1][j-wt[i-1]][k-vol[i-1]],
                        table[i-1][j][k]
                    )
                else:
                    table[i][j][k] = table[i-1][j][k]

    return table[n][w][v]

print("Bottom-up DP:", DP_knapSack(W, V, wt, vol, val, n))



#Memoization
calc = [[[-1]*(V+1) for _ in range(W+1)] for _ in range(n+1)]

def mem_knapSack(w, v, wt, vol, val, n):
    if n == 0 or w == 0 or v == 0:
        return 0
    if calc[n][w][v] != -1:
        return calc[n][w][v]

    if wt[n-1] <= w and vol[n-1] <= v:
        calc[n][w][v] = max(
            val[n-1] + mem_knapSack(w - wt[n-1], v - vol[n-1], wt, vol, val, n-1),
            mem_knapSack(w, v, wt, vol, val, n-1)
        )
    else:
        calc[n][w][v] = mem_knapSack(w, v, wt, vol, val, n-1)

    return calc[n][w][v]

print("Top-down DP:", mem_knapSack(W, V, wt, vol, val, n))
```

Fibonacci Numbers

```python
def fibonacci(n):
    sequence = []
    
    if n >= 1:
        sequence.append(0)
    if n >= 2:
        sequence.append(1)
    
    for i in range(2, n):
        next_number = sequence[i-1] + sequence[i-2]
        sequence.append(next_number)
    
    return sequence

n = 10
fib_seq = fibonacci(n)

print(f"The first {n} Fibonacci numbers are:")
for num in fib_seq:
    print(num, end=" ")
```

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
#type your code here
def fibonacci_n(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1

    fib = [0] * (n + 1)
    fib[0] = 0
    fib[1] = 1

    for i in range(2, n + 1):
        fib[i] = fib[i-1] + fib[i-2]

    return fib[n]

n = 10
nth_fib = fibonacci_n(n)
print(f"The {n}th Fibonacci number is: {nth_fib}")
```

## Supplementary Problem (HOA 2.1 Submission):
* Choose a real-life problem
* Use recursion and dynamic programming to solve the problem

```python
#type your code here for recursion programming solution
def min_days_recursive(topics):
    if topics == 0:
        return 0
    if topics < 0:
        return 1000  #use big number instead of infinity

    return 1 + min(
        min_days_recursive(topics - 1),
        min_days_recursive(topics - 3),
        min_days_recursive(topics - 5)
    )

n = 7
days = min_days_recursive(n)
print("Recursion: Minimum days to complete", n, "topics is", days)
```

```python
#type your code here for dynamic programming solution
def min_days_dp(topics):
    dp = [1000] * (topics + 1)
    dp[0] = 0

    for i in range(1, topics + 1):
        if i >= 1:
            dp[i] = min(dp[i], 1 + dp[i - 1])
        if i >= 3:
            dp[i] = min(dp[i], 1 + dp[i - 3])
        if i >= 5:
            dp[i] = min(dp[i], 1 + dp[i - 5])

    return dp[topics]

n = 7
days = min_days_dp(n)
print("DP: Minimum days to complete", n, "topics is", days)
```
