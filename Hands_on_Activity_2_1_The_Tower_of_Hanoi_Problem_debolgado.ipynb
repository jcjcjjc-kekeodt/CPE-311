```python
n = 4
srce= 'A'
auxi = 'C'
desti = 'B'

def move_disks(n, srce, auxi, desti):
    if n == 1:
        print(f'Move disk 1 from rod {srce} -> rod {desti}')
        return

    move_disks(n - 1, srce, desti, auxi)
    print(f'Move disk {n} from rod {srce} -> rod {desti}')
    move_disks(n - 1, auxi, srce, desti)

#Stage 1:
print("STAGE 1:")
print("SOURCE: A")
print("AUXILIARY: B")
print("DESTINATION: C")
print("**MOVE:**")

move_disks(n, srce, auxi, desti)



#Stage 2:
print("STAGE 2:")
print("SOURCE: B")
print("AUXILIARY: A")
print("DESTINATION: C")
print("**MOVE:**")
move_disks(n, srce, auxi, desti)


#Stage 3:
print("STAGE 3:")
print("SOURCE: A")
print("AUXILIARY: B")
print("DESTINATION: C")
print("**MOVE:**")
move_disks(n, srce, auxi, desti)

#Biggest in A to C
print("BIGGEST IN A TO C")
print("SOURCE: A")
print("AUXILIARY: C")
if n == 1:
    print(f'Move disk 1 from rod {srce} -> rod {desti}')

move_disks(n-1, srce, desti, auxi)
print(f'Move disk {n} from rod {srce} -> rod {desti}')
move_disks(n - 1, auxi, srce, desti)
```
