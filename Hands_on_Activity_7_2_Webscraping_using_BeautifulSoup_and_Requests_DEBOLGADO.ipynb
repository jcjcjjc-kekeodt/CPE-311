## **Example of gathering image data using webcam** <br>
Note: Run this snippet using local jupyter notebook

```python
import cv2
from google.colab.patches import cv2_imshow

# Start webcam
webcam = cv2.VideoCapture(0)

while True:
    try:
        check, frame = webcam.read()
        print(check)   # True as long as webcam is running
        print(frame)   # Matrix values of each frame

        cv2.imshow("Capturing", frame)
        key = cv2.waitKey(1)

        if key == ord('s'):
            cv2.imwrite(filename='saved_img.jpg', img=frame)
            webcam.release()

            img_new = cv2.imread('saved_img.jpg', cv2.IMREAD_GRAYSCALE)
            cv2.imshow("Captured Image", img_new)
            cv2.waitKey(1650)
            cv2.destroyAllWindows()

            print("Processing image...")
            img_ = cv2.imread('saved_img.jpg', cv2.IMREAD_ANYCOLOR)
            print("Converting RGB image to grayscale...")
            gray = cv2.cvtColor(img_, cv2.COLOR_BGR2GRAY)
            print("Converted RGB image to grayscale...")

            print("Resizing image to 28x28 scale...")
            img_ = cv2.resize(gray, (28, 28))
            print("Resized...")

            cv2.imwrite(filename='saved_img-final.jpg', img=img_)
            print("Image saved!")

            break
        elif key == ord('q'):
            print("Turning off camera.")
            webcam.release()
            print("Camera off.")
            print("Program ended.")
            cv2.destroyAllWindows()
            break

    except KeyboardInterrupt:
        print("Turning off camera.")
        webcam.release()
        print("Camera off.")
        print("Program ended.")
        cv2.destroyAllWindows()
        break
```

```python
# import required libraries
import sounddevice as sd
from scipy.io.wavfile import write
import wavio as wv

# Sampling frequency
freq = 44100

# Recording duration
duration = 5

# Start recorder with the given values
# of duration and sample frequency
recording = sd.rec(int(duration * freq),
                   samplerate=freq, channels=2)

# Record audio for the given number of seconds
sd.wait()

# This will convert the NumPy array to an audio
# file with the given sampling frequency
write("recording0.wav", freq, recording)

# Convert the NumPy array to audio file
wv.write("recording1.wav", recording, freq, sampwidth=2)
```

#**Web Scraping of Movies Information using BeautifulSoup**
We want to analyze the distributions of IMDB and Metacritic movie ratings to see if we find anything interesting. To do this, weʼll first scrape data for over 2000 movies.


```python
import requests
from bs4 import BeautifulSoup

url = "https://asurascanz.com/?2026-02-26"

response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

manhwas = soup.find_all("div", class_="bs")  # bs listing blocks

for m in manhwas:
    # Title
    title_tag = m.find("div", class_="bsx")
    title = title_tag.find("a")["title"] if title_tag else "N/A"

    # Genre
    genre_tag = m.find("div", class_="limit")
    genre = genre_tag.get_text(strip=True) if genre_tag else "N/A"

    # update date
    date_tag = m.find("span", class_="date")
    update_date = date_tag.get_text(strip=True) if date_tag else "N/A"

    print(f"Title: {title}, Genre: {genre}, Updated: {update_date}")
```

**Using Beautiful soups and requests**


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

url = "https://asurascanz.com/?2026-02-26"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

os.makedirs("covers", exist_ok=True)

data = []

manhwas = soup.find_all("div", class_="bs")

for idx, m in enumerate(manhwas[:10]):  # limit to first 10
    # Title
    title_tag = m.find("div", class_="bsx")
    title = title_tag.find("a")["title"] if title_tag else "N/A"

    # Genre
    genre_tag = m.find("div", class_="limit")
    genre = genre_tag.get_text(strip=True) if genre_tag else "N/A"

    # Update date
    date_tag = m.find("span", class_="date")
    update_date = date_tag.get_text(strip=True) if date_tag else "N/A"

    # Cover image
    img_tag = m.find("img")
    img_url = img_tag["src"] if img_tag else None

    if img_url:
        img_data = requests.get(img_url).content
        filename = f"covers/{title.replace(' ', '_')}.jpg"
        with open(filename, "wb") as f:
            f.write(img_data)

    data.append([title, genre, update_date, img_url])
df = pd.DataFrame(data, columns=["Title", "Genre", "Updated", "Cover URL"])

print(df)
```

**Using Selenium**


```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import requests
import pandas as pd
import os
options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=options)
driver.get("https://asurascanz.com/?2026-02-26")

time.sleep(5)

os.makedirs("covers", exist_ok=True)

data = []

manhwas = driver.find_elements(By.CSS_SELECTOR, "div.bs")

for idx, m in enumerate(manhwas[:10]):  # limit to first 10
    # Title
    title = m.find_element(By.CSS_SELECTOR, "div.bsx a").get_attribute("title")

    # Genre/description snippet
    genre = m.find_element(By.CSS_SELECTOR, "div.limit").text

    # Update date
    update_date = m.find_element(By.CSS_SELECTOR, "span.date").text

    # Cover image
    img_url = m.find_element(By.CSS_SELECTOR, "img").get_attribute("src")

    if img_url:
        img_data = requests.get(img_url).content
        filename = f"covers/{title.replace(' ', '_')}.jpg"
        with open(filename, "wb") as f:
            f.write(img_data)

    data.append([title, genre, update_date, img_url])

driver.quit()
df = pd.DataFrame(data, columns=["Title", "Genre", "Updated", "Cover URL"])

print(df)
```
