```python
import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/Meteorite_Landings.csv'
df = pd.read_csv(file_path)

print(df.head())

```

# STEP 2

```python
# keep only the first 4 characters
df['year'] = df['year'].str.slice(0, 4)

# convert to numeric
df['year'] = pd.to_numeric(df['year'], errors='coerce')

print(df[['year']].head(10))

```

# STEP 3

```python
df['observed_before_1970'] = df['year'] < 1970

print(df[['year', 'observed_before_1970']].head(10))
```

# STEP 4

```python
df = df.set_index('id').sort_index()

print(df.head())
```

# STEP 5

```python
subset = df.loc[10036:10040]
print(subset)
```

# BONUS

```python
# look for suspicious years "less than 1000 or greater than current year"
errors = df[(df['year'] < 1000) | (df['year'] > 2026)]
print(errors[['year']])
```
