```python
import pandas as pd

file_path = '/content/drive/My Drive/Colab Notebooks/Meteorite_Landings.csv'
df = pd.read_csv(file_path)

df['year'] = df['year'].str.slice(0,4)
df['year'] = pd.to_numeric(df['year'], errors='coerce')
```

```python
# filter years 2005â€“2009
df_filtered = df[(df['year'] >= 2005) & (df['year'] <= 2009)]

# pivot table: count + 95th percentile of mass
pivot = pd.pivot_table(
    df_filtered,
    index='year',
    columns='fall',   # "Found" vs "Fell"
    values='mass (g)',
    aggfunc={'mass (g)': ['count', lambda x: x.quantile(0.95)]}
)

print(pivot)
```

```python
# summary statistics for Found vs Fell
summary_stats = df.groupby('fall')['mass (g)'].describe()
print(summary_stats)
```

# EXERCISE PART 5

```python
taxi_file = '/content/drive/My Drive/Colab Notebooks/2019_Yellow_Taxi_Trip_Data.csv'
taxi = pd.read_csv(taxi_file, parse_dates=['tpep_dropoff_datetime'])

# Set datetime index
taxi = taxi.set_index('tpep_dropoff_datetime')
```

```python
hourly = taxi.resample('H').agg({
    'trip_distance': 'sum',
    'fare_amount': 'sum',
    'tolls_amount': 'sum',
    'tip_amount': 'sum'
})

print(hourly.head())
```

```python
top5_tips = hourly.sort_values('tip_amount', ascending=False).head(5)
print(top5_tips)
```
